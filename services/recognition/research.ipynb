```python
embedding1 = DeepFace.represent(img_path = "../../tmp_files/Photo on 17.11.2021 at 08.21.jpg", model_name = 'Facenet')
embedding2 = DeepFace.represent(img_path = "../../tmp_files/user_photo.jpg", model_name = 'Facenet')
embedding3 = DeepFace.represent(img_path = "dataset/img1.jpg", model_name = 'Facenet')

embedding1, embedding2, embedding3

```

```python
import numpy as np

#distance between two images

distance_vector = np.square(np.array(embedding1) - np.array(embedding2))
#print(distance_vector)

distance = np.sqrt(distance_vector.sum())
print("Euclidean distance: ",distance)

#distance between two images

distance_vector = np.square(np.array(embedding2) - np.array(embedding3))
#print(distance_vector)

distance = np.sqrt(distance_vector.sum())
print("Euclidean distance: ",distance)
```
